```python

"""
 created by deepak 
 description importing the libraries used for the movie data analysis
"""
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%pylab inline
```

```python
"""
created by deepak 
 description data frame for reading all the movie list created and seeing the top data on the list using the head function
"""
movie_df=pd.read_csv('tmdb-movies.csv')
movie_df.head()
```

```python
"""
created by deepak
description dataframe to give the mean of movies according to the release year.
Here we are trying to group all the movies and group them according to their release year and finally taking the mean of the data.
"""
data_movie=movie_df.groupby(['release_year'],as_index=False).mean()
data_movie.head()
```

# Standarized Function
This method takes the list and calculates its mean, standard deviation and returns the list with the standarized values in it

```python
def standard_list(list_to_standardize):
    """
    created by deepak
    description the method taked the list and send the list after standardizing it by performing the vectorized operations to it.
    """
    mean=list_to_standardize.mean()
    sd=list_to_standardize.std()
    return (list_to_standardize-mean)/sd
```

Here we are creating a one dimension list named as gain list where we are storing the profit made by each movie by taking out the difference between the revenue and the budget of the movie we have already omitted the outliers from the list.

```python
"""
created by deepak
description The total gain of the movies by taking the difference of revenue and the budget of the movies using the describe method
to decribe the list data.
"""
gain_list=movie_budget_df['revenue']-movie_budget_df['budget']
gain_list.describe()
```

Here we are calling the standarized function created above and passing the gain list to know how much standard deviation away the revenue of each movie is from the list the list shows the profit in the standard deviation from the mean profit made by all movies.
This list can be used in various analysis. As of now the it is not used in the analysis.

```python
standard_list(gain_list)
```

# Hypothesis Testing of popularity of the action movie
Here in this hypothesis test we will try to prove the null hypothesis wrong that the popularity of action movies is different from the normal movies by performing the two tail z-test on our data set. The sample size of our data set is more than 30 and we will assume that we are considering the total population and not the sample data of the movies.

H n

```python
def action_movie_list(genre_df):
    action_series=pd.Series([])   
    action_df = pd.DataFrame(columns=['vote_average'])
    for index, row in genre_df.iterrows():        
        genre_string=row['genres']               
        if genre_string!=NaN:
            #movie_list=pd.Series(genre_string.split('|'))
            movie_list=np.array(genre_string.split('|'))
            for i in np.nditer(movie_list):
                if(i=='Action'):
                    action_df=action_df.append({'vote_average':row['vote_average']}, ignore_index=True) 

    return action_df
```

```python

genre_df=movie_df[movie_df.genres>0]
action_movie_df=action_movie_list(genre_df)
population_mean=genre_df['vote_average'].mean()
population_std=genre_df['vote_average'].std()
action_movie_mean=action_movie_df['vote_average'].mean()
print 'The mean of the population is {} , the std of population is {} and the mean of the action movies are {}'.format(population_mean,population_std,action_movie_mean)
z_score=(action_movie_mean-population_mean)/population_std
print z_score
```
